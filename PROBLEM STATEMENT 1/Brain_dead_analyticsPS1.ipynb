```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
datasheet=pd.read_csv('State_wise_rice_production.csv')
X=datasheet.iloc[:-1, 1:].values
df = pd.DataFrame(X)

```

```python
df=df.replace('.',np.nan)
```

```python
df=df.replace('-',np.nan)
```

```python
print(df)
```

```python

string_values = df.applymap(lambda x: isinstance(x, str))
print(string_values)
```

```python
df = df.astype(float)
```

```python

string_values = df.applymap(lambda x: isinstance(x, str))

print(string_values)
```

```python
df = df.apply(lambda row: row.fillna(row.mean()), axis=1)
```

```python
print(df)
```

```python

df['Total_Production'] = df.sum(axis=1)
```

```python
print(df)
```

```python
plt.figure(figsize=(12, 6))
df['Total_Production'].plot(kind='bar', color='skyblue')
plt.title('Total State-wise Rice Production(2004-2023)')
plt.xlabel('State/Union Territory')
plt.ylabel('Production (in metric tons)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```

```python
growth_rates = df.pct_change(axis=1) * 100 
```

```python
print(growth_rates)
```

```python
average_production_rate = growth_rates.iloc[:, 1:-1].mean(axis=1)
```

```python
print(average_production_rate)
```

```python
import matplotlib.pyplot as plt


colors = plt.cm.coolwarm(range(len(average_production_rate)))

plt.figure(figsize=(12, 6))
plt.bar(average_production_rate.index, average_production_rate, color=colors)
plt.title('Average Growth Rate by State/Union Territory (2004-2023)', fontsize=16)
plt.xlabel('State/Union Territory', fontsize=12)
plt.ylabel('Average Production Rate (%)', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```
